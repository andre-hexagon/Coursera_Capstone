# The Battle of Neighborhoods

## Introduction | Business Problem

When people are trying to open a new business, they face multiple key decisions before the business even starts to run that will determine the success or failure of the enterprise. These decisions range from the business category Most frequently, the businessmen take such decisions based on intuition or gut feeling

This project will focus on leveraging the location data, as well as the resources from Foursquare to help people that are looking forward to open a new business in the city make the best decision based on the business the founder would like to establish.

This will help them answer the following questions:
1. Which neighborhood has the most # of business of the same category?
2. What are the main competitors?
3. What would be the best latitude, longitude based on the business around the neighborhood of the same category?
4. What is the centroid of the cluster of businesses on the chosen category?

## Data
